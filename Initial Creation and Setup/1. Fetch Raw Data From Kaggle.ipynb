```python
import kagglehub
from kagglehub import KaggleDatasetAdapter
import os
import pandas as pd
import shutil

VOLUME_RAW_PATH = "/Volumes/zillow/zillow_medallion/raw"

# Download dataset locally and get directory path
dataset_path = kagglehub.dataset_download("zillow/zecon")

print("Dataset downloaded to:", dataset_path)

# List all files in dataset
all_files = []
for root, dirs, files in os.walk(dataset_path):
    for file in files:
        all_files.append(os.path.join(root, file))

print("Files found:")
for f in all_files:
    print(f)

```

```python
# Download dataset (KaggleHub cache)
dataset_path = kagglehub.dataset_download("zillow/zecon")

print("Downloaded to cache:", dataset_path)

# Move all files directly into UC Volume
for root, dirs, files in os.walk(dataset_path):
    for file in files:
        source_file = os.path.join(root, file)
        target_file = os.path.join(VOLUME_RAW_PATH, file)

        # Copy file into Volume
        shutil.copy2(source_file, target_file)
        print(f"Copied: {file} â†’ {VOLUME_RAW_PATH}")
```
