```python
RAW = "/Volumes/zillow/zillow_medallion/raw"
ASIS_DIR = "/Volumes/zillow/zillow_medallion/raw-converted_format/as_is"

dbutils.fs.rm("dbfs:" + ASIS_DIR, True)
dbutils.fs.mkdirs("dbfs:" + ASIS_DIR)

files = [
  "City_time_series.csv",
  "Metro_time_series.csv",
  "CountyCrossWalk_Zillow.csv",
  "cities_crosswalk.csv",
  "DataDictionary.csv",
  "all_available_metrics.json",
  "fields_per_level.json"
]

for f in files:
    src = f"dbfs:{RAW}/{f}"
    dst = f"dbfs:{ASIS_DIR}/{f}"
    dbutils.fs.cp(src, dst, True)
    print("Copied:", f)

display(dbutils.fs.ls("dbfs:" + ASIS_DIR))
```
