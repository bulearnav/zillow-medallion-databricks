```python
import dlt
from pyspark.sql import functions as F

# Use the folder, not the file
SRC_DIR = "/Volumes/zillow/zillow_medallion/raw-converted_format/dlt"

@dlt.table(
  name="neighborhood_ts_dlt_bronze",
  comment="Neighborhood time series ingested via DLT using Auto Loader (CSV)."
)
def neighborhood_ts_dlt_bronze():
    return (
        spark.readStream.format("cloudFiles")
          .option("cloudFiles.format", "csv")
          .option("cloudFiles.inferColumnTypes", "true")  # better than inferSchema here
          .option("header", "true")
          .load(SRC_DIR)
          .withColumn("load_dt", F.current_timestamp())
          .withColumn("source_file", F.input_file_name())
          .withColumn("ingest_mode", F.lit("dlt_autoloader_csv"))
    )
```
