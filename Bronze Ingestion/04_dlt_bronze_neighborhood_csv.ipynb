```python
import dlt
from pyspark.sql import functions as F

# Best practice: Use cloudFiles for robust CSV ingestion in DLT
SRC = "/Volumes/zillow/zillow_medallion/raw-converted_format/dlt/Neighborhood_time_series.csv"

@dlt.table(
  name="neighborhood_ts_dlt_bronze",
  comment="Neighborhood time series ingested to Bronze using Delta Live Tables."
)
def neighborhood_ts_dlt_bronze():
    return (
        spark.readStream.format("cloudFiles")
          .option("cloudFiles.format", "csv")
          .option("header", "true")
          .option("inferSchema", "true")
          .load(SRC)
          .withColumn("load_dt", F.current_timestamp())
          .withColumn("source_file", F.lit(SRC))
          .withColumn("ingest_mode", F.lit("dlt_csv"))
    )
```
